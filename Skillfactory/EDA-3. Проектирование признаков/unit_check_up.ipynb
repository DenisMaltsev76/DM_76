```python
import pandas as pd
df = pd.read_csv('wine_cleared.csv')
df.head()
```

```python
df.info()
```

```python
# для удобства сразу преобразуем признак в int
df['price_round'] = df['price'].round().astype(int)
df.head()
```

```python
regex = '\d{4}' # регулярное выражение для нахождения чисел
df['year'] = df['title'].str.findall(regex).str.get(0)
df.head()
```

```python
# создать новый бинарный признак is_usa и присвоить ему 1 в случае, если вино произведено в США, иначе — 0
df['is_usa'] = df['country'].apply(lambda x: 1 if x == 'US' else 0)
df.head()
```

```python
# Определить три самых часто встречающихся страны
top_countries = df['country'].value_counts().head(3)

# Вывести результаты
print("Три самых часто встречающихся страны в столбце 'country':")
print(top_countries)
```

```python
# создать новый бинарный признак is_usa и присвоить ему 1 в случае, если вино произведено в Италии, иначе — 0
df['is_Italy'] = df['country'].apply(lambda x: 1 if x == 'Italy' else 0)

# создать новый бинарный признак is_usa и присвоить ему 1 в случае, если вино произведено во Франции, иначе — 0
df['is_France'] = df['country'].apply(lambda x: 1 if x == 'France' else 0)

df['is_France'].sum()
```
