### <center> Цикл while

Цикл while подразумевает немного другой подход к ограничению количества шагов, которые должны выполняться.

Цикл выполняется до тех пор, пока истинно его условие. Как только оно становится ложным, цикл прерывается.

Например, мы разрабатываем новую умную колонку «Алевтина». В ней заложен алгоритм воспроизведения любимой песни, пока пользователь не скажет одно из ключевых слов («Стоп», «Хватит», «Помолчи» и т. д.). Таким образом, здесь заложен цикл воспроизведения музыки, условие выхода из него — ключевое слово пользователя.

Для компьютера это выглядит как: «Делай, пока не наступит ...» .

Условие — это логическое выражение, которое может принимать значение True и False. Например, условие first > second, будет означать, что цикл будет выполняться до тех пор, пока значение переменной first больше значения переменной second. Если значение условия становится равным False, цикл перестаёт выполняться.

Для работы с циклом while используется следующая конструкция:

while условие:  
    # Начало блока кода с телом цикла   
    # Пока условие истинно, цикл выполняется    
    ...     
    ...     
    ...     
    # Конец блока кода с телом цикла    
#Код, который будет выполняться после цикла     
Такой цикл ещё называют циклом с предусловием.

Давайте рассмотрим структуру цикла while на примере.    

Пусть мы хотим удалять элементы в списке lst=['a', 'b', 'c', 'd', 'e'], пока их не останется два.   
Тогда синтаксис цикла для решения нашей задачи будет иметь следующий вид:   

```python
lst = ['a', 'b', 'c', 'd', 'e']
# Создаём цикл с условием, что длина списка больше 2.
while len(lst) > 2: #делай, пока…
    # Начало блока кода с телом цикла
    
    # Обновляем список, удаляя первый элемент в нём.
    lst.pop(0) 
    # Выводим промежуточный результат
    print(lst) 


    #Конец блока кода с телом цикла


# Выводим на экран результирующий список
print('Result', lst)
```

#####  Задача №1

Условие задачи. Пассажирский лифт имеет ограничение на перевозку — не более 400 кг единовременно. Пусть входящие в лифт люди имеют одинаковый вес (weight).

Необходимо написать программу, которая должна следить за изменением нагрузки на лифт, сравнивать её с грузоподъёмностью и в случае перевеса выдавать предупреждение 'Overweight N kg'). Например, при перевесе в 15 кг должно быть выведено сообщение 'Overweight 15 kg').

Для данного примера вес заходящих людей зададим как weight = 67).

Решение

По традиции обсуждаем алгоритм решения задачи.

Итак, мы заранее не знаем, сколько людей сможет уехать в лифте, но зато знаем условие перегрузки (остановки цикла): суммарный вес не должен превышать 400 кг. Отсюда делаем вывод, что нам подойдёт цикл while. Помимо переменных weight (вес человека) и max_weight (грузоподъёмность), создадим переменную S, в которой будем считать суммарный вес людей в лифте. До начала цикла S = 0. 

Воспользуемся нашим алгоритмом создания цикла while:

Задаём условие работы цикла — суммарный вес S меньше грузоподъемности max_weight.
Определяем действия в цикле — в цикле будем увеличивать суммарный вес S на величину веса входящего в лифт человека weight. Дополнительно будем выводить текущее значение суммарного веса S после его обновления.
По окончании работы цикла будем выводить на экран получившийся перевес — разницу между суммарным весом S и грузоподъёмностью max_weight.

Оборачиваем наши рассуждения в код:

```python
# Задаём вес входящего в лифт человека
weight = 67 
# Задаём грузоподъёмность
max_weight = 400 
# Задаём суммарный вес людей в лифте
S = 0 
# Создаём цикл, который будет работать, пока S не превысит max_weight. 
while S < max_weight: # делай, пока…
    # Увеличиваем суммарный вес
    # Равносильно S = S + weight 
    S += weight 
    # Выводим значение суммарного веса после обновления
    print('Current sum weight', S) 


# Отделяем промежуточный вывод от результата пустой строкой
print() 
# Выводим итоговое значение перевеса
print('Overweight {} kg'.format(S - max_weight))
```

##### Задача №2

Условие задачи. Написать цикл, который будет складывать натуральные числа, пока их сумма не превысит 500.

Натуральные числа — это числа 1, 2, 3, 4 и т. д.

Решение

Обсудим алгоритм решения данной задачи. Мы заранее не знаем число шагов нашего цикла, но знаем условие, при котором нужно остановиться. Поэтому выбираем цикл while. Создадим две переменные для суммы S и для текущего числа n. До начала цикла сумма S = 0, а первое число n = 1.

Итак, согласно нашему алгоритму создания цикла while, мы должны:

Определить условие работы цикла — цикл должен выполняться, пока сумма S меньше 500.
Задать действие в теле цикла — увеличить переменную S (рассчитывать сумму натуральных чисел) на значение переменной n. Увеличить текущее число n на 1 (1, 2, 3, 4…). Дополнительно будем выводить строку 'Still counting...' (справочная строка, которая говорит, что идёт подсчёт).
По окончании работы цикла выведем результирующую сумму и количество чисел, которые нам понадобилось сложить, чтобы сумма превысила пороговое значение 500. Количество чисел вычисляется как итоговое значение n после цикла минус 1.

Наконец, реализуем наш алгоритм в виде кода:

```python
# Создаём накопительную переменную, в которой будем считать сумму.
S = 0  
# Задаём текущее натуральное число
n = 1  
 
# Создаём цикл, который будет работать, пока сумма не превысит 500.
while S < 500:  # делай, пока …
    # Увеличиваем сумму, равносильно S = S + n
    S += n
    # Увеличиваем значение натурального числа  
    n += 1  
    # Выводим строку ожидания
    print("Still counting...") 
# Отделяем промежуточный вывод от результата пустой строкой
print() 
# Выводим результирующую сумму
print("Sum is: ", S) 
# Выводим результирующее количество чисел
print("Numbers total: ", n-1)
```

##### Задача №3

Условие задачи. Представьте себя администратором тайного интернет-сообщества «Три Слепые Мыши» с ограниченным членством. У каждого участника сообщества есть свой позывной и пароль. Позывные и пароли представлены в словаре secret_passwords. Ключи данного словаря — позывные, а значения — пароли.

secret_passwords = {
    'Enot': 'ulybaka',
    'Agent12': '1password1',
    'MouseLulu': 'myshkanaruhka'
}
Необходимо создать бесконечный цикл, который будет проверять, что пользователь, посетивший интернет-сообщество, является его членом. Пользователь вводит свои позывной и пароль с клавиатуры.

Решение

Перейдём к обсуждению алгоритма решения данной задачи.

Итак, по условию задачи нам необходимо в бесконечном цикле запрашивать у пользователя позывной (назовём его переменной name) и пароль к нему (назовём его переменной password) и проверять, есть ли информация о пользователе с такими данными в нашей системе. Таким образом, нам нужно проверить, есть ли указанный позывной среди ключей словаря secret_passwords, а затем проверить, совпадает ли введённый пароль со значением по введённому ключу.

Посмотрим на наш алгоритм создания цикла while:

Определяем условие выполнения цикла — это будет значение True, то есть цикл будет бесконечным. Условие остановки такого цикла — совпадение введённого пароля с паролем в словаре secret_passwords.
Задаём действие в цикле — запросить с помощью функции input() у пользователя его позывной, проверить, есть ли указанный позывной среди ключей словаря secret_passwords. Если позывной незнаком системе, будем выводить сообщение 'Wrong name'. Если позывной совпадает, то у пользователя запрашивается пароль. Если пароль не совпадает, то выводится сообщение 'Wrong password'. Если пароль совпадает со значением в словаре, выводим сообщение 'Welcome' и завершаем цикл.  
Оформляем наш алгоритм в виде кода:

```python
# База позывных и паролей
secret_passwords = {
    'Enot': 'ulybaka',
    'Agent12': '1password1',
    'MouseLulu': 'myshkanaruhka'
} 
# Создаём бесконечный цикл
while True:
    # Запрашиваем у пользователя позывной
    name = input('Enter your name: ') 
    # Проверяем, что позывной есть среди ключей словаря.
    if name in secret_passwords: 
        # Если позывной верный, запрашиваем у пользователя пароль.
        password = input('Enter your password: ')
        # Проверяем, что введённый пароль совпадает со значением по ключу позывного.
        if password == secret_passwords[name]: 
            # Если пароль верный, выводим приветственное сообщение.
            print('Welcome')
            # Завершаем цикл
            break 
        else: 
            # Если пароль неверный, выводим сообщение об ошибке.
            print('Wrong password')
    else:
        # Если позывной неверный, выводим сообщение об ошибке.
        print('Wrong name')
```

На каждой итерации цикла в переменную name заносится значение, которое вводится с клавиатуры с помощью функции input(). В операторе if проверяется условие, что значение переменной name принадлежит набору ключей словаря secret_passwords.

Если это условие не выполняется, происходит переход к блоку внешнего else, где выводится сообщение об ошибочном позывном, тело цикла завершается, и происходит переход на новую итерацию. 
Если name принадлежит набору ключей словаря secret_passwords, то в переменную password заносится значение, которое вводится с клавиатуры с помощью функции input().
В операторе if проверяется равенство значения переменной password и значения словаря, которое лежит по ключу name.

Если равенство не выполняется, то происходит переход к блоку внутреннего else, где выводится ошибка об ошибочном пароле, тело цикла завершается, и происходит переход на новую итерацию.
Если равенство выполняется, то выводится приветственное сообщение и цикл завершает свою работу.

##### Задание 3.4

Напишите бесконечный цикл while с условием выхода внутри цикла, который находит минимальное натуральное число, квадрат которого больше value.

Запишите результат в переменную number.

```python
number = 1
value = int(input('Enter your value:'))
while True:
    if number ** 2 > value:
        break
    number += 1
print(number)
```

##### Задание 3.11

Олег положил money рублей в банк под 8 % годовых с ежегодной капитализацией процентов. Через сколько лет у него на счету будет не менее target_money рублей?

Запишите ответ в переменную year_count.

Для моделирования процесса увеличения денежной суммы на счёте воспользуйтесь циклом while. В случае, когда значение target_money < money, в переменной year_count должен содержаться 0.

```python

money = 1000
target_money = 3000
year_count = 0        # Начальное значение количества лет
while money < target_money:
    money *= 1.08    # Увеличиваем сумму на 8% с ежегодной капитализацией
    year_count += 1  # Увеличиваем количество лет на 1
print(year_count)
```
