## <center> Вложенные циклы

Представим, что у нас есть некоторый датчик измерения температуры, который записывает температуру воздуха в градусах Цельсия три раза в сутки — утром, днём и вечером. Пусть температура утром была 13 градусов, днём — 15, а вечером — 9. Все три наблюдения за один день мы можем оформить в виде списка, например [13, 15, 9].

Но что нам делать, если мы измеряем температуру несколько дней подряд, например три дня? Заводить на каждый день отдельную переменную со списком? А если мы измеряем температуру в течение длительного периода — допустим, два года?

Получается, нам нужно создать более 600 переменных на каждый день! Конечно же, никто так не делает. В таком случае используют вложенные списки. То есть мы создаём один список, в котором храним списки измерений температуры на каждый день. Например:

```python
temperature = [
    [13, 15, 10],
    [14, 13, 9],
    [8, 9, 6]
]
```

Такие структуры напоминают таблицу, и их ещё называют двумерными списками, а на математическом языке они называются двумерными матрицами.

Двумерная матрица представляет собой обычную таблицу. Если одномерная матрица — это список, то двумерная — это список списков.

Как и у обычной таблицы, у нашего двумерного списка есть условные строки и столбцы. 

Строками таблицы назовём списки, которые находятся внутри одного большого списка. Например, второй (а если говорить «по-питоновски», то первой) строкой нашей таблицы является список [14, 13, 9] — наблюдения за второй день.

Элементы, находящиеся в строках таблицы, образуют условные столбцы таблицы.

Чтобы обратиться к третьему измерению температуры за второй день, нам надо сначала извлечь наблюдения за второй день — это список, который хранится под индексом 1 в списке temperature, 

```python
temperature = [
    [13, 15, 10],
    [14, 13, 9],
    [8, 9, 6]
]
print(temperature[1])
## Будет выведено:
## [14, 13, 9]
```

Теперь нам осталось только достать третье по счёту измерение (вечерний замер температуры). Для этого обращаемся к результату по индексу 2:

```python
temperature = [
    [13, 15, 10],
    [14, 13, 9],
    [8, 9, 6]
]
print(temperature[1][2])
## Будет выведено
## 9
```

Переводя на язык таблиц: для доступа к элементу необходимо сначала обратиться по индексу строки, а затем — по индексу столбца.

В дальнейшем мы часто будем использовать общее выражение «элемент в i-й строке и j-м столбце». Это будет означать, что мы рассматриваем элемент, который стоит под индексами i и j, где i и j — это переменн

Примечание. Стоит отметить, что в строгом математическом смысле матрица имеет ограничение на размерности её строк. Не может быть, чтобы в одной из строк матрицы было четыре элемента, а в другой — шесть. В то же время во вложенном списке такая ситуация вполне возможна и не вызовет ошибки.
